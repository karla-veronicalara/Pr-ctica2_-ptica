```python
x=np.linspace(-2*np.pi, 2*np.pi,150)
omega=np.array([1,2,7,32,15,np.pi])

for i in range(len(omega)):
    y=np.sinc(omega[i]*x)
    plt.plot(x,y, label=r'$\omega$ = %2.2f' %omega[i])

plt.legend(loc=2)
plt.show()
```

Derivada asimtrica
$$
f'_i=\frac{f_{i+1}}{}  
$$

```python
def deriv_Euler(f,x):
    #NEcesito un arreglo donde poner las dericadas
    df = np.zeros_like(f)
    for i in range(len(x)-1):
        df[i]=(f[i+1]-f[i])/(x[i+1]-x[i])
    df[-1]=-(f[-2]-f[-1])/(x[-1]-x[-2])
    return df

def deriv_sim(f,x):
    #NEcesito un arreglo donde poner las dericadas
    df = np.zeros_like(f)
    for i in range(len(x)-1):
        df[i]=(f[i+1]-f[i])/(x[i+1]-x[i])
    df[0]=(f[1]-f[0])/(x[1]-x[0])
    df[-1]=-(f[-2]-f[-1])/(x[-1]-x[-2])
    return df
    
```

```python
x=np.linspace(0,2*np.pi)
y=np.sin(x)
dy_a=np.cos(x)
dy_n= deriv_Euler(y,x)
dy_s=deriv_sim(y,x)

plt.plot(x,y,label=r'f(x)')
plt.plot(x,dy_a,label=r'f`(x) analtica')
plt.plot(x,dy_n,label=r'f`(x) numricA')

plt.legend()
plt.show()

#modificaciones pendientes en foto 1
```

```python
def deriv_h2(f,x):
    #NEcesito un arreglo donde poner las derivadas
    df = np.zeros_like(f)
    for i in range(len(x)-2):
        df[i]=-(-1.5*f[i] + 2*f[i+1] - 0.5*f[i+2])/(x[i+1]-x[i])
    df[-2]=-(-1.5*f[-2] + 2*f[-3] - 0.5*f[-4])/(x[-2]-x[-3])
    df[-1]=-(-1.5*f[-1] + 2*f[-2] - 0.5*f[-3])/(x[-1]-x[-2])
    return df
```
