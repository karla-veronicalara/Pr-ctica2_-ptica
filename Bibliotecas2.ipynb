```python
def E_q(r,r_q,q):
    d32=((r[0]-r_q[0])**2 + (r[1]-r_q[1])**2)**1.5
    return (q*(r[0]-r_q[0]))/d32, (q*(r[1]-r_q[1]))/d32
```

```python
x,y = np.linspace(-2,2) , np.linspace(-2,2)
XX,YY = np.meshgrid(x,y)

#Esta N es para normalizar mis vectores
N = np.sqrt(XX**2 + YY**2)

plt.quiver(XX,YY,XX/N,-YY/N)
plt.gca().set_aspect('equal')
```

```python
#Funci√≤n para n cargas 
def E_tot(r,r_qs,qs):
    Ex,Ey = np.zeros_like(r[0]), np.zeros_like(r[1])
    for i in range(len(qs)):
        E= E_q(r,r_qs[i],qs[i])
        Ex= Ex + E[0]
        Ey= Ey + E[1]
    return Ex,Ey
```

```python
pos_qs = [[0,1],[0,1]]
qs = [1,-1]
Ex, Ey = E_tot([XX,YY],pos_qs,qs)
plt.streamplot(XX,YY,Ex,Ey,density=2)
```
